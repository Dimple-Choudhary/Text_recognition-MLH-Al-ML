```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_rgb(img):
  plt.figure(figsize=(12,10))
  return plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
